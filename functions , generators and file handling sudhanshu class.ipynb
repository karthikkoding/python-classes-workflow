```python
def genfib(n):
    a=1
    b=1
    output=[]
    for i in range(n):
        output.append(a)
        a,b = b,a+b
    return output
```

```python
def genfib(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
genfib(10)
```

```python
for i in genfib(1070):
    print(i)
```
